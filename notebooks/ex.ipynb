# Assignment 5 - Model 

```python
# load dataset
from pycaret.datasets import get_data
juice = get_data('juice')

# init setup
from pycaret.classification import *
exp_name = setup(data = juice,  target = 'Purchase')

# train model
lr = create_model('lr')
```

```python
# Confusion Matrix
plot_model(lr, plot = 'confusion_matrix', plot_kwargs = {'percent' : True})
```

```python
# Precision-Recall Curve
plot_model(lr, plot = 'pr', plot_kwargs = {'percent' : True})
```

```python
# ROC Curve
plot_model(lr, plot = 'auc', plot_kwargs = {'percent' : True})
```

```python
# Classification Report
plot_model(lr, plot = 'class_report', plot_kwargs = {'percent' : True}, )
```
